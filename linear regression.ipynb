# defining the functions here

```python
#line equation
def line(m,x,c):
    return m*x+c
#error function
def error(m,x,c,MPG):
    return np.mean((line(m,x,c)-MPG)**2)
def derivative_slope(m,x,c,MPG):
    return 2*(np.mean(line(m,x,c)-MPG)*x)
def derivative_intercept(m,x,c,MPG):
    return 2*(np.mean(line(m,x,c)-MPG ))







```

# using gradient descent to reduce the error and get exact m,c values

```python
m=random.random()
c=random.random()
cost=[]
alpha=0.000001
for i in range (0,600): 
    m=m-alpha*derivative_slope(m,feature,c,target_var)
    c=c-alpha*derivative_intercept(c,m,feature,target_var)
    cost.append(error(m,feature,c,target_var))
plt.plot(cost)
print("m=:",m)
print("C=:",c)
    
```
